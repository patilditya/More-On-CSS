```python
# Question 5 : Create a simple blog website about Google fonts. The webpage must have the heading with “Oswald” font align center to the page, below the heading have an image of Google fonts and a paragraph about Google fonts in font “Montserrat” paragraph font style should be“montserrat” font&

--> HTML File :

   <!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Google Fonts Blog</title>
    <!-- Link to Google Fonts -->
    <link href="https://fonts.googleapis.com/css2?family=Oswald:wght@400&family=Montserrat:wght@400&display=swap" rel="stylesheet">
    <style>
        /* Apply the Oswald font to the heading */
        h1 {
            font-family: 'Oswald', sans-serif;
            text-align: center;
        }
        /* Apply the Montserrat font to the paragraph */
        p {
            font-family: 'Montserrat', sans-serif;
            font-size: 18px;
            line-height: 1.6;
            text-align: center;
            margin: 20px;
        }
        /* Style for the image */
        img {
            display: block;
            margin: 0 auto;
            width: 80%;
            max-width: 600px;
        }
        /* Add some margin for the body */
        body {
            margin: 20px;
            font-family: sans-serif;
        }
    </style>
</head>
<body>

    <h1>Exploring Google Fonts</h1>

    <!-- Image of Google Fonts -->
    <img src="https://www.gstatic.com/images/branding/googlelogo/2x/googlelogo_light_color_92x30dp.png" alt="Google Fonts Logo">

    <!-- Paragraph about Google Fonts -->
    <p>Google Fonts is a collection of free and open-source fonts that are optimized for the web. It offers hundreds of different styles, allowing developers and designers to easily access and implement beautiful typography in their projects. You can choose from an extensive library of fonts, and easily add them to your website with a simple link or CSS import.</p>

</body>
</html>
```
